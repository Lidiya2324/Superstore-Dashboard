#### 4.2 Handle Missing Values

```python
missing = df.isnull().sum()

missing_percent = (missing / len(df)) * 100

pd.DataFrame({
    "Missing Values": missing,
    "Percentage": missing_percent
})
```

The dataset contained **24 missing values** in the `Income` column (approximately **1.07%** of all records).

Before imputing these values, the distribution of the `Income` variable was examined using a box plot and the Interquartile Range (IQR) method to determine whether outliers were present. Since the distribution contained extreme values, the **median** was chosen instead of the mean because it is less sensitive to outliers.

```python
df["Income"] = df["Income"].fillna(df["Income"].median())
```

```python
df.isnull().sum()
```

#### 4.3 Detect and Handle Outliers

##### Detect and Evaluate Outliers

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df['Income'])
plt.show()
```

A box plot of the Income variable indicated the presence of several potential outliers. The IQR method was then used to identify observations outside the acceptable range.

```python
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Income'] < lower_bound) | (df['Income'] > upper_bound)]

print("Number of outliers:", len(outliers))
outliers[['Income']]
```

##### Interpretation

The IQR method identified 8 potential outliers in the Income variable.

* Seven customers reported annual incomes between 153,924 and 162,397. Although these values are high, they are relatively close to one another and therefore appear to represent a valid group of high-income customers rather than data entry errors.

* One customer reported an annual income of 666,666, which is substantially higher than all other observations. This record was investigated further before deciding whether any preprocessing was required.

```python
df.loc[2233]
```

The customer with an annual income of 666,666 exhibited purchasing behavior that was inconsistent with other high-income customers. Despite reporting the highest income in the dataset, this customer made relatively few purchases and spent very little across all product categories. This observation suggests that the income value may be anomalous or the result of a data entry error.

```python
df = df[df["Income"] != 666666]
```

Seven high-income observations were retained because they represented plausible customer records. One observation with an annual income of 666,666 was removed after further investigation indicated that its purchasing behaviour was inconsistent with the reported income, suggesting a potential anomaly.

#### 4.6 Standardize Text Values

The categorical columns were inspected for inconsistent capitalization, extra spaces, and uncommon category labels.

The `education` and `marital_status` columns were standardized by removing leading/trailing whitespace and applying title case. Although the values were already consistently formatted, this step ensures the dataset follows a uniform standard.

Additionally, the rare categories `Alone`, `Absurd`, and `YOLO` in the `marital_status` column were consolidated into `Single` to improve consistency while preserving all customer records.

```python
print(df["education"].unique())

print(df["marital_status"].unique())
```

```python
df["marital_status"].value_counts()
```

```python
# Standardize text formatting
df["education"] = df["education"].str.strip().str.title()

df["marital_status"] = df["marital_status"].str.strip().str.title()
```

```python
df["marital_status"] = df["marital_status"].replace({
    "Alone": "Single",
    "Absurd": "Single",
    "Yolo": "Single"
})
```

```python
print(df["education"].unique())
print(df["marital_status"].value_counts())
```
